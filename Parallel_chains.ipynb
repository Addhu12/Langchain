### FINAL CHAIN = MAIN CHAIN + PARALLEL-CHAIN 1 + PARALLEL-CHAIN 2

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnableLambda


# 'MAIN CHAIN'

# TASK 1: Defining "PROMPT TEMPLATE"
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a movie summarizer"),
    ("human", "Please summarize the movie in brief: {input}")
])


# TASK 2: Defining "MODEL"
llm_openai = ChatOpenAI(model="gpt-4o-mini", temperature=0)


# TASK 3: Defining "PARSER"
str_parser = StrOutputParser()

# TASK 4: Defining "RUNNABLE"
def dictionary_maker(text:str)-> dict:

    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)



# Step 5: "PARALLEL CHAIN -1"
def linkedin_chain(text: dict):
    text = text["text"]
    
    linkedin_prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a linkedin post generator"),
        ("human", "Create a post for: {text}")
    ])
    chain_linkedin = linkedin_prompt | llm_openai | str_parser

linkedin_chain_runnable = RunnableLambda(linkedin_chain)



# Step 6: "PARALLEL CHAIN -2"
def insta_chain(text: dict):
    text = text["text"]
    
    insta_prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an Instagram post generator"),
        ("human", "Create an Instagram post: {text}")
    ])

    chain_insta = insta_prompt | llm_openai | str_parser

insta_chain_runnable = RunnableLambda(insta_chain)






# Step 7: "FINAL CHAIN"
final_chain = (
                prompt_template |
                llm_openai |
                str_parser |
                dictionary_maker_runnable |
                RunnableParallel(
                        branches = {
                                "linkedin": chain_linkedin, 
                                "instagram": chain_insta
                        }
                )
)

# Step 8: Invoke
final_chain.invoke("KGF")
```

### CHAIN AS A "RUNNABLE" (Can do this too)

```python
# I)  "BEAUTIFY" FUNCTION


from langchain_core.runnables import RunnableLambda

def beautify(final_response:dict) -> dict:

    linkedin_response = final_response['branches']['linkedin']
    insta_response = final_response['branches']['instagram']

    return {"linkedin": linkedin_response, "instagram": insta_response}


beautify_runnable = RunnableLambda(beautify)


# II) "FINAL CHAIN"


from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnableLambda

# 'MAIN CHAIN'

# TASK 1: Defining "PROMPT TEMPLATE"
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a movie summarizer"),
    ("human", "Please summarize the movie in brief: {input}")
])

# TASK 2: Defining "MODEL"
llm_openai = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# TASK 3: Defining "PARSER"
str_parser = StrOutputParser()

# TASK 4: Defining "RUNNABLE"
def dictionary_maker(text: str) -> dict:
    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)


# Step 5: "PARALLEL CHAIN - 1" (Defined globally)
linkedin_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a linkedin post generator"),
    ("human", "Create a post for: {text}")
])
chain_linkedin = linkedin_prompt | llm_openai | str_parser


# Step 6: "PARALLEL CHAIN - 2" (Defined globally)
insta_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an Instagram post generator"),
    ("human", "Create an Instagram post: {text}")
])
chain_insta = insta_prompt | llm_openai | str_parser


# Step 7: "FINAL CHAIN"
final_chain = (
    prompt_template |
    llm_openai |
    str_parser |
    dictionary_maker_runnable |
    RunnableParallel(
        branches = {
            "linkedin": chain_linkedin, 
            "instagram": chain_insta
        }
    )
)

# Step 8: Invoke
final_chain.invoke("KGF")









# III) BEAUTIFIED CHAIN

beautified_chain = final_chain | beautify_runnable

beautified_chain.invoke("Kurup")








```
