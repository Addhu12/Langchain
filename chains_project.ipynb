## CHAINS with "Custom Runnable"

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

print(os.getenv("OPENAI_API_KEY"))

llm_openai = ChatOpenAI(model = "gpt-4o-mini", temeperature = 0)
```

```python
# TASK 1: Creating Prompt

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_prompts([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# TASK 2: Sending message to LLM

llm_openai = ChatOpenAI(model = "gpt-4o-mini", temperature = 0)
```

```python
# TASK 3: Parser

# It will only fetch the content (Automating) -- String output parser
# Need to understand the libraries (like langchain_core.outputparsers)
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

```python
# TASK 3: Custom function

from langchain_core.runnables import RunnableLambda
def dictionary_maker(text: str) -> dict:
    return {"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)

# We need to apply a wrapper on top of our function, so that chain in langchain will treat our function as runnable
```

```python
# TASK 5 [Template for "POST"]

prompt_post = ChatPromptTemplate.from_messages(
    messages = [
        ("system", "You are a social media post generator."),
        ("human", "Create a post for the following text for LinkedIn: {text}")
    ]
)

```

```python
# TASK 6[Generate the Post]

llm_openai = ChatOpenAI(model_name = "gpt-4o-mini", temperature = 0)
```

```python
# TASK 7[String Output Parser]

str_parser = StrOutputParser()
```

```python
# CHAIN

chain = prompt_template | llm_openai | str_parser | dictionary_maker_runnable | prompt_post | llm_openai | str_parser 

chain.invoke("What is the capital of France?")
```
