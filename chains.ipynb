```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

print(os.getenv("OPENAI_API_KEY"))

llm_openai = ChatOpenAI(model = "gpt-4o-mini", temeperature = 0)


```

#### First Chain

```python
# TASK 1: Creating Prompt

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_prompts([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])

```

```python
# TASK 2: Sending message to LLM

llm_openai = ChatOpenAI(model = "gpt-4o-mini", temperature = 0)
```

```python
# TASK 3: Parser

# It will only fetch the content (Automating) -- String output parser
# Need to understand the libraries (like langchain_core.outputparsers)
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()

```

#### Manual Invocation

```python
## template = prompt_template.invoke({"input": "What is the capital of France?"})

## res = llm.invoke(template)

## final_result = res.content
```

#### Chain Invocation

```python
chain = prompt_template | llm_openai | str_parser

chain.invoke({"input": "What is the capital of France?"})
```

```python
from langchain_core import output_parsers
from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(prompt_template, llm_openai, output_parser)

chain_1.invoke("What is the capital of France?")





```
