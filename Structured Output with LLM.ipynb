## USING "PYDANTIC"

```python
# Code for "Joke setup"

# Providing precise context

from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
load_dotenv()

print(os.getenv("OPENAI_API_KEY"))

llm_openai = ChatOpenAI(model = "gpt-4o-mini", temperature = 0)

from pydantic import BaseModel, Field
class llm_schema(BaseModel):
    setup: str = Field(description = "The setup for the joke")
    punchline: str = Field(description = "The Punchline for the joke")


```

```python

llm_structured_output = llm_openai.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me a joke")
# type(result)  ## Type of result is "llm schema"

result.punchline


```

## Just additional

```python
# Creating "PYDANTIC Object"

obj = llm_schema(**{"setup": "some setup", "punchline": " some punchline"})
obj

# If LLM generates an output & doesn't follow a schema, it will not allow LLM to generate the output & protect us to generate pseudo-schema or broken schema to downstream workflows --> DATA VALIDATION

# This is a kind of "Data validation" checking ; It also checks "data type"

# DATA PARSING
obj = llm_schema(**{"setup": 123, "punchline": "some punchline"}) 
obj
```
